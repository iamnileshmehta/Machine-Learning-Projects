```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
```

```python
df= pd.read_csv('/content/iris1.csv')
df.head()
```

```python
df.describe()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
sepal_mean= df['sepal width (cm)'].mean()
print(sepal_mean)
petal_mean= df['petal length (cm)'].mean()
print(petal_mean)

df['sepal width (cm)'].fillna(sepal_mean, inplace=True)
df['petal length (cm)'].fillna(petal_mean, inplace= True)
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder

lb= LabelEncoder()
df['Class']= lb.fit_transform(df['Class'])
df
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(df)
plt.grid()
```

```python
Q1_sepal= df['sepal width (cm)'].quantile(0.25)
Q3_sepal= df['sepal width (cm)'].quantile(0.75)

iqr_sepal= Q3_sepal-Q1_sepal
print(iqr_sepal)
lower_bound_sep = Q1_sepal-1.5 *iqr_sepal
upper_bound_sep = Q3_sepal +1.5 *iqr_sepal

```

```python
Q1_pet= df['petal length (cm)'].quantile(0.25)
Q3_pet= df['petal length (cm)'].quantile(0.75)

iqr_pet= Q3_pet-Q1_pet
print(iqr_pet)

lower_bound_pet= Q1_pet -1.5*iqr_pet
upper_bound_pet= Q3_pet +1.5*iqr_pet

df1= df[(df['sepal width (cm)']>= lower_bound_sep) & (df['sepal width (cm)']<= upper_bound_sep) &
 (df['petal length (cm)'] >= lower_bound_pet) & (df['petal length (cm)']<= upper_bound_pet)]
```

```python
df1
```

```python
df1.describe()
```

```python
df1.shape
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(df1)
plt.grid()
```

```python
correlation= df1.corr()
sns.heatmap(correlation, annot= True)
```

Split the DataSets

```python
from sklearn.model_selection import train_test_split

X= df1.drop(columns=['Class'])
y= df1['Class']

X_train, X_test, y_train, y_test= train_test_split (X, y, test_size=0.20, random_state= 7)

print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
from sklearn.model_selection import cross_val_score, KFold #machine learning algorithms
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier

models= []
models.append(('lr',LogisticRegression()))
models.append(('decision tree',DecisionTreeClassifier()))
models.append(('svm',SVC(gamma='auto')))
models.append(('knn',KNeighborsClassifier()))
models.append(('naive bayes',GaussianNB()))
models.append(('Random Forest',RandomForestClassifier()))

for name,model in models:
  kfold=KFold(n_splits=10,random_state=7, shuffle= True)
  cross_val_sc=cross_val_score(model,X,y,scoring='accuracy',cv=kfold)
  print('{} : acc: {}(standard deviation: {})'.format(name,cross_val_sc.mean(),cross_val_sc.std()))
```

Train the Model

```python
model= LogisticRegression()
model.fit(X_train, y_train)
y_pred= model.predict(X_test)
print(y_pred)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```
